```python
import sys,json
import plotly.graph_objects as go

import pandas as pd

from datetime import datetime
```

```python
fpath = '/content/drive/MyDrive/assinatura/tabela_assinatura_completo.csv'


assinatura = pd.read_csv(fpath,parse_dates=['log_timestamp'])



```

```python
assinatura
```

```python
#preparação dos dados

def preprocess(df):

  df.drop(['id','log_message','ip',"cpfPessoa","idPessoa","ipSistema","ipUsuario","username","usuarioServico","casOAuthScope", "emissor","operacao","serialNumber","hash"], axis=1, inplace=True)
  
  df.rename({'log_timestamp':'Data',
           'casClientId':'Cliente',
           'status':'Status'}, axis=1, inplace=True)
  
  #'YYYY-[M]M-[D]D [[H]H:[M]M:[S]S]'
  df['Data'] = df['Data'].dt.strftime('%Y-%m-%d')
  
  
  return df

assinatura = preprocess(assinatura)
```
